## LinearRegression

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('AmesHousing.txt', sep = '\t')
```

```python
X = data[['Gr Liv Area']]
y = data['SalePrice']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
lr = LinearRegression()
#train the model.
lr.fit(X_train, Y_train)
```

```python
predictionsY = lr.predict(X_test)
predictionsY
```

### Mean squared error

To quantify the fit, we can use mean squared error.

```python
from sklearn.metrics import mean_squared_error
print("Mean squared error: %.2f" % mean_squared_error(Y_test, predictionsY))
```

```python
test_rmse = mean_squared_error(Y_test, predictionsY)
test_rmse
```

```python
import numpy as np
test_sq_rmse = np.sqrt(test_rmse)
test_sq_rmse
```
